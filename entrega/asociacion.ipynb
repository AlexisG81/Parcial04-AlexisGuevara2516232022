```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Convertir TODOS los valores a texto

df = df.astype(str)

# Convertir a lista de transacciones

transacciones = df.values.tolist()
```

```python
te = TransactionEncoder()

te_array = te.fit(transacciones).transform(transacciones)

df_apriori = pd.DataFrame(
    te_array,
    columns=te.columns_
)

print(df_apriori.head())
```

```python
df_apriori.head()
```

```python


frequent_itemsets = apriori(
    df_apriori,
    min_support=0.05,
    use_colnames=True
)

print(frequent_itemsets.head())
```

## Aplicación del algoritmo Apriori

Se utilizó el algoritmo Apriori para encontrar patrones frecuentes de compra entre productos.

El soporte mínimo se configuró en 5%, permitiendo identificar combinaciones relevantes dentro de las transacciones.

```python
reglas = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.5
)

print(reglas.head())
```

## Generación de reglas de asociación

Las reglas de asociación fueron generadas utilizando:

- Soporte
- Confianza
- Lift

Estas métricas permiten medir la fuerza de relación entre productos.

```python
top_reglas = reglas.sort_values(
    by='lift',
    ascending=False
)

top10 = top_reglas.head(10)

top10
```

```python
top10[[
    'antecedents',
    'consequents',
    'support',
    'confidence',
    'lift'
]]
```

DESCRIPCION DEL TOP 10

Las reglas fueron ordenadas utilizando la métrica lift, permitiendo identificar las relaciones más fuertes entre productos.
